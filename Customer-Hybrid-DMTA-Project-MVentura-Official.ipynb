# Task 1.2 - Age best predictor for Credit Card eligibility? Yes

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X_train_age, X_test_age, y_train_age, y_test_age = train_test_split(df[['Age']], df.Credit_Card, test_size=0.2, random_state=42)
X_train_creditscore, X_test_creditscore, y_train_creditscore, y_test_creditscore = train_test_split(df[['Credit_Score']], df.Credit_Card, test_size=0.2, random_state=42)
X_train_balance, X_test_balance, y_train_balance, y_test_balance = train_test_split(df[['Balance']], df.Credit_Card, test_size=0.2, random_state=42)
X_train_yearsemployed, X_test_yearsemployed, y_train_yearsemployed, y_test_yearsemployed = train_test_split(df[['Years_Employed']], df.Credit_Card, test_size=0.2, random_state=42)
X_train_accountlength, X_test_accountlength, y_train_accountlength, y_test_accountlength = train_test_split(df[['Account_Length']], df.Credit_Card, test_size=0.2, random_state=42)
```

```python
model_age = LogisticRegression(class_weight='balanced')
model_age.fit(X_train_age, y_train_age)

model_creditscore = LogisticRegression(class_weight='balanced')
model_creditscore.fit(X_train_creditscore, y_train_creditscore)

model_balance = LogisticRegression(class_weight='balanced')
model_balance.fit(X_train_balance, y_train_balance)

model_yearsemployed = LogisticRegression(class_weight='balanced')
model_yearsemployed.fit(X_train_yearsemployed, y_train_yearsemployed)

model_accountlength = LogisticRegression(class_weight='balanced')
model_accountlength.fit(X_train_accountlength, y_train_accountlength)
```

```python
y_pred_age = model_age.predict(X_test_age)
print("Accuracy for Age:", accuracy_score(y_test_age, y_pred_age))
print("Classification Report for Age:")
print(classification_report(y_test_age, y_pred_age))

y_pred_creditscore = model_creditscore.predict(X_test_creditscore)
print("Accuracy for Credit Score:", accuracy_score(y_test_creditscore, y_pred_creditscore))
print("Classification Report for Credit Score:")
print(classification_report(y_test_creditscore, y_pred_creditscore))

y_pred_balance = model_balance.predict(X_test_balance)
print("Accuracy for Balance:", accuracy_score(y_test_balance, y_pred_balance))
print("Classification Report for Balance:")
print(classification_report(y_test_balance, y_pred_balance))

y_pred_yearsemployed = model_yearsemployed.predict(X_test_yearsemployed)
print("Accuracy for Years Employed:", accuracy_score(y_test_yearsemployed, y_pred_yearsemployed))
print("Classification Report for Years Employed:")
print(classification_report(y_test_yearsemployed, y_pred_yearsemployed))

y_pred_accountlength = model_accountlength.predict(X_test_accountlength)
print("Accuracy for Account Length:", accuracy_score(y_test_accountlength, y_pred_accountlength))
print("Classification Report for Account Length:")
print(classification_report(y_test_accountlength, y_pred_accountlength))
```
